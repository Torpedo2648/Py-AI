```python
import numpy as np
import matplotlib.pyplot as plt
import math
import networkx as nx
from turtle import color
```

```python
G = nx.Graph()
G.clear()
G.add_nodes_from([1,2,3,"four",(5, {"color":"red"})])
G.add_edges_from([(1,"four"),(1,3),(3,5),(5,"four"),(2,5),(2,3),(1,2),(2,"four")])

fig, ax = plt.subplots()
nx.draw(G=G,ax=ax, with_labels=True, node_color=range(5))

print(list(nx.connected_components(G)))
print(list(nx.shortest_path(G)))
```
